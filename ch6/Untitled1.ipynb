```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25,
            4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 2],
    '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10,
           29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
    '광고비': [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642,
            591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
    '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893,
            830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
print(df.head(3))
```

```python
print('1:',round(df.할인율.corr(df.온도),2))

model=ols('주문량~광고비+온도+할인율',df).fit()
print('2:',round(model.rsquared,2))

print('3:',round(model.params,4))

print('4:',round(model.pvalues['온도'],4))
print('통계적으로 유의하다')

new_data=pd.DataFrame({'할인율':[10],'온도':[20],'광고비':[500]})
pred=model.predict(new_data)
print('6:',pred)

print('7:',sum(model.resid**2))

print('8:',(model.resid**2).mean())


print('9:',model.conf_int(alpha=0.1).loc['온도'])
new_data=pd.DataFrame({'할인율':[15],'온도':[25],'광고비':[300]})
pred=model.get_prediction(new_data)
print(pred.summary_frame(alpha=0.1))


model=ols('주문량~광고비',df).fit()

if model.pvalues['광고비']<0.05:
    print('귀무가설기각')
else:
    print('채택')
```

```python
a=df[:int(len(df)*0.5)]
b=df[int(len(df)*0.5):]

from statsmodels.formula.api import logit
model=logit('target~age+service+social+booked',a).fit()

# print(model.summary())
print(sum(model.pvalues[1:]>=0.05))
# social 유의하지않음 1개


model=logit('target~age+booked',a).fit()   
print(model.pvalues.sort_values(ascending=False))

print(abs(model.params).sort_values(ascending=False))

print(model.llf)

print(-2*model.llf)

import numpy as np

print(np.exp(3*model.params['booked']))

print(model.params[model.pvalues<0.05].sum())

pred=model.predict(b)
pred=(pred>0.5).astype('int')
from sklearn.metrics import accuracy_score
print(accuracy_score(b['target'],pred))

ac=accuracy_score(b['target'],pred)

print(1-ac)
```
