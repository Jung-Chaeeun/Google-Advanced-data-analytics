```python
import pandas as pd
# df = pd.read_csv("math.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part3/ch6/math.csv")
print(df.head())
```

```python
groupa=df['groups']=='group_A'
groupb=df['groups']=='group_B'
groupc=df['groups']=='group_C'
groupd=df['groups']=='group_D'

print(stats.shapiro(df[groupa]['scores']))
print(stats.shapiro(df[groupb]['scores']))
print(stats.shapiro(df[groupc]['scores']))
print(stats.shapiro(df[groupd]['scores']))


print(stats.levene(df[groupa]['scores'],df[groupb]['scores'],df[groupc]['scores'],df[groupd]['scores']))
```

```python
from statsmodels.formula.api import ols

model=ols('scores ~ groups',df).fit()

from statsmodels.stats.anova import anova_lm
anova_lm(model)
```

```python
df=pd.read_csv('customer_travel.csv')
df
```

```python
dfa=df[:400]
dfb=df[400:]
dfb
```

```python
model=logit('target ~ age+service+social+booked',dfa).fit()
model.summary()
```

```python
model2=logit('target ~ age+booked',dfa).fit()
model2.summary()
```
